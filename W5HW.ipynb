1. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a normal distribution and your prior for $\tau$ is a gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
import pymc as pm
import arviz as az
import numpy as np
import matplotlib.pyplot as plt

# sample of normal data
np.random.seed(424)
n_samples = 30
true_theta = 0
true_tau = 2
data = np.random.normal(loc=true_theta, scale=1/np.sqrt(true_tau), size=n_samples)

# PyMC model
normal_gamma_model = pm.Model()

with normal_gamma_model:
    # Priors
    theta0 = 0  # theta belief
    theta_prior_n = 1  # strength of prior belief as units of data
    tau0 = theta_prior_n * true_tau  # precision

    theta = pm.Normal("theta", mu=theta0, sigma=1/np.sqrt(tau0))
    tau = pm.Gamma("tau", alpha=2, beta=2)

    # likelihood as observed value
    x_obs = pm.Normal("x_obs", mu=theta, sigma=1/np.sqrt(tau), observed=data)

# Run MCMC sampling
m = 5000
with normal_gamma_model:
    idata = pm.sample(draws=m, chains=2, tune=100)

# Extract posterior samples
theta_samples = idata.posterior["theta"].values
tau_samples = idata.posterior["tau"].values

az.plot_trace(idata)
plt.show()

az.summary(idata)
```

2. *Use PyMC to provide Bayesian inference for the paramaters associated with a sample of normal data where your prior for theta is a non normal distribution and your prior for $\tau$ is a non-gamma distribution. Provide diagnostic assessments of the performance of your algorithm.*

```python
np.random.seed(424)
n_samples = 30
true_theta = 0
true_tau = 2
data = np.random.normal(loc=true_theta, scale=1/np.sqrt(true_tau), size=n_samples)

model2 = pm.Model()
m = 5000
with model2:
    # Priors (uniform for theta, exponential for tau)
    theta = pm.Uniform("theta", lower=-2, upper=2)
    tau = pm.Exponential("tau", lam=1)

    x_obs = pm.Normal("x_obs", mu=theta, sigma=1/np.sqrt(tau), observed=data)

    idata = pm.sample(draws=m, chains=2, tune=100)

theta_samples = idata.posterior["theta"].values
tau_samples = idata.posterior["tau"].values

az.plot_trace(idata)
plt.show()

az.summary(idata)
```

### Q3: Slice Sampling

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant. In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*

```python
def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):

    # find interval of grid points where f(x_grid) > y
    # then extend the enterval so endpoints f(a)<y and f(b)<y
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid

    x_ = a + stats.uniform().rvs()*(b-a)
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f"
    elif x_ < x:  # or if "x_ was above f on the left side of the interval"
        x_l,x_r = x_,b
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):

    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)

    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1)
```

```python
from scipy import stats

x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))

m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):

    x[t,1] = stats.uniform().rvs()*f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')

    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1])
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')

plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```

Slice sampling, a Markov Chain Monte Carlo algorithm, samples by selecting a random horizontal slice at y under the density function and then drawing a uniform sample from the area where the function is above that slice. The code initializes a starting point, and a probability distribution function f(x). For each iteration, a random height y is chosen between 0 and f(x),  creating a horizontal slice under the function curve. The algorithm then selects a new x\_ from the interval above y (where f(x)>y). If the new x\_ satsifies f(x_)>y, it is accepted; otherwise, the interval is shrunk and the process repeats until we get an accepted one. This draws samples proportionally to the target distribution without requiring a proposal like Metropolis-Hastings (MH).

In a Metropolis within Gibbs algorithm, slice sampling can replace the MH step where the full conditionals are only known up to a normalizing constant. Instead of using the MH proposal to generate a new value and accepting it with a probability based on a proposal distribution, slice sampling chooses the sampling interval with the process above. The curve being sampled is the full conditional density of the parameter, with the y slice being the valid region for sampling. The initial value in Gibbs sampling is simply the previous sample value, and the sample value is then used in the Gibbs sampler update. This means we avoid risking a high rejection rate at MH, since the slicing guarantees we properly explore the joint posterior distribution.
