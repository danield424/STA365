## Bayes Theorem

\begin{align*}
\overset{posterior}{p(\theta|x)} &={} \frac{\overset{likelihood}{p(x|\theta)}\times \overset{prior}{p(\theta)}}{\overset{\text{evidence}}{p(x)}}\\
&={} \frac{{p(x, \theta)}}{\underset{\text{marginal likelihood}}{\displaystyle \int p(x|\theta)p(\theta) d\theta}} \textrm{ or }
\frac{{p(x, \theta)}}{\underset{\text{law of total probability}}{\displaystyle \sum_{k} p(x|\theta_k)p(\theta_k)}} =
\frac{\overset{joint}{p(x, \theta)}}{\underset{\text{normalizing constant}}{c}}
\end{align*}

### _The posterior is always proportional to the joint_

$$\large p(\theta|x) = (1/c)\times p(\theta,x) \propto{} p(\theta,x)$$

**_and the posterior is always a proper (integrates to 1) distribution if the prior is_**

Q1: What proportion of students who answer this question quickly will pass the class?
Assume the probability of passing the class is 0.9.
Assume the probability of answering this question quickly is 0.6 if you'll pass the class,
while the probability drops to 0.3 if you'll not pass the class.

$P(Pass|Quick=1) = \frac{P(Quick = 1|Pass) * P(Pass)}{P(Quick = 1)} = \frac{P(Quick = 1|Pass) * P(Pass)}{P(Quick = 1|Pass) * P(Pass)  +  P(Quick = 1|¬¨Pass) * (1- P(Pass))}$

$= \frac{0.6*0.9}{0.6*0.9 + 0.3*0.1} = \frac{0.54}{0.57} \approx 0.95$

Approximately 95% of students who answer the question quickly will pass the class.

\end{align*}

The posterior is created by combining the likelihood, which is a multinomial distribution, and the prior, which is a Dirichlet distribution:

- **Likelihood**:
  $p(x|\theta, n) = \text{Multinomial}(x_1, x_2, \dots, x_k | n, \theta) = \frac{n!}{x_1! x_2! \cdots x_k!} \prod_{i=1}^k \theta_i^{x_i}$

- **Prior**:
  $p(\theta|\alpha) = \text{Dirichlet}(\theta_1, \theta_2, \dots, \theta_k | \alpha_1, \alpha_2, \dots, \alpha_k) = \frac{1}{B(\alpha)} \prod_{i=1}^k \theta_i^{\alpha_i - 1}$
  where $ B(\alpha)$ is the normalizing constant of the Dirichlet distribution.

- **Posterior**: Using Bayes' Theorem:
$p(\theta|x, n, \alpha) = \frac{\left( \frac{n!}{x_1! x_2! \cdots x_k!} \prod_{i=1}^k \theta_i^{x_i} \right) \left( \frac{1}{B(\alpha)} \prod_{i=1}^k \theta_i^{\alpha_i - 1} \right)}{p(x)}$

Combine terms in the numerator:

$p(\theta|x, n, \alpha) = \frac{\frac{n!}{x_1! x_2! \cdots x_k! B(\alpha)} \prod_{i=1}^k \theta_i^{x_i + \alpha_i - 1}}{p(x)}$

Now we drop normalizing constants $\frac{n!}{x_1! x_2! \cdots x_k!}, B(\alpha)$, and $p(x)$ since we only care about the distribution shape.

$p(\theta|x, n, \alpha) \propto \prod_{i=1}^k \theta_i^{x_i + \alpha_i - 1}$

The posterior distribution is a Dirichlet distirbution
$p(\theta|x, n, \alpha) = \text{Dirichlet}(\theta_1, \theta_2, \dots, \theta_k | \alpha_1', \alpha_2', \dots, \alpha_k'),
$
with updated parameters:
$\alpha_i' = \alpha_i + x_i, \quad \text{for } i = 1, 2, \dots, k.$

Q3: Create a sample from a three category multinomial distribution and visualize the multinomial distribution and the multinomial sample; then, visualize the simplex Dirichlet distribution of the prior and the posterior using a ternary plot.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import plotly.figure_factory as ff

### Sample from a three-category multinomial distribution
n = 500 #trials
true_pbs = np.array([0.5, 0.3, 0.2])  # True probabilities for the multinomial
np.random.seed(424)
sample = np.random.multinomial(n, true_pbs, size=1)
normalized_sample = sample / n

# Visualize the multinomial distribution and sample
categories = ['p1', 'p2', 'p3']

# Bar plot for the sample counts
plt.bar(categories, sample.flatten(), color='skyblue', edgecolor='black')
plt.title('Counts from Multinomial Sample (n = 500)')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.show()

# Bar plot for distribution's true probabilities and sample probabilities
x = np.arange(len(categories))  # X-axis positions
bar_width = 0.3
plt.bar(x - bar_width / 2, true_pbs, width=bar_width, label='True Probabilities', color='blue')
plt.bar(x + bar_width / 2, normalized_sample.flatten(), width=bar_width, label='Sample Probabilities', color='lightblue')

plt.xticks(x, categories)
plt.xlabel('Categories')
plt.title('Multinomial Distribution and Sample')
plt.ylabel('Probability')
plt.legend()
# print(normalized_sample)
plt.show()

### Simplex Dirichlet distribution of the prior
alpha_prior = np.array([1.5, 1.5, 1.5])
# This choice represents a slightly non-uniform prior over the simplex.
# The values 1.5>1 bias the prior slightly toward balanced probabilities (e.g., [1/3,1/3,1/3]), but not too strongly. alpha going the other way would skew towards extremes
dirichlet_prior = stats.dirichlet(alpha=alpha_prior)

# Define a grid for the simplex, with 100 evenly spaced possible values
p1 = np.linspace(0, 1, 100)
p2 = 1 - p1.copy()
p_ = np.linspace(0, 1, 100)
# Create a mesh grid for the probabilities
p1 = (p1[np.newaxis, :] * p_[:, np.newaxis]).flatten()
p2 = (p2[np.newaxis, :] * p_[:, np.newaxis]).flatten()
p3 = 1 - p1 - p2
# Combine p1, p2, p3 into a 2D array where each row is a probability vector
p = np.c_[p1, p2, p3]
p = p[p[:, 2] >= 0]  # Remove invalid probabilities

# Calculate dirichlet pdf for each point: p.T transposes the grid to match the expected input shape of the PDF function (columns represent the categories).
dirichlet_prior_densities = dirichlet_prior.pdf(p.T)

fig_prior = ff.create_ternary_contour(
    p.T,
    dirichlet_prior_densities,
    pole_labels=["p1", "p2", "p3"],
    interp_mode="cartesian",
    showscale=True,
    colorscale="YlOrRd",  # Use Plotly's built-in yellow-to-red gradient
)
fig_prior.update_layout(title="Dirichlet Prior Distribution") #integrating this will give 1.
fig_prior.show('png')
print(f"Prior parameters (alpha values): {alpha_prior}")

### Simplex Dirichlet distribution of the posterior using the sample
alpha_posterior = alpha_prior + sample.flatten() # essentially counts + prior
dirichlet_posterior = stats.dirichlet(alpha=alpha_posterior)
dirichlet_posterior_densities = dirichlet_posterior.pdf(p.T)
fig_posterior = ff.create_ternary_contour(
    p.T,
    dirichlet_posterior_densities,
    pole_labels=["p1", "p2", "p3"],
    interp_mode="cartesian",
    showscale=True,
    colorscale="YlOrRd",
)
fig_posterior.update_layout(title="Dirichlet Posterior Distribution") #peaks around the observed data.
fig_posterior.show('png')
print(f"Posterior parameters (alpha values): {alpha_posterior}")
```
